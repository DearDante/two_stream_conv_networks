```python
import cv2
import matplotlib.pyplot as plt
```

```python
video = cv2.VideoCapture('v_YoYo_g25_c05.avi')
```

```python
img = video.read()[1]
b, g, r = cv2.split(img)
img2 = cv2.merge([r,g,b])

plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img2)
```
